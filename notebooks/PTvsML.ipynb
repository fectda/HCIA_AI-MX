# Problema visto desde la perspectiva de programación tradicional

```python
def normalRule(x):
    return 2*x-1
```

```python
#Datos que ya teníamos
print(normalRule(-1))
print(normalRule(0))
print(normalRule(1))
print(normalRule(2))
```

```python
#Datos desconocidos
print(normalRule(10))
print(normalRule(45))
```

# Problema visto desde la perspectvia de Machine Learning (DL)

```python
import tensorflow as tf
import numpy as np
```

```python
input_x = np.array([-1,0,1,2,3,4],dtype=float)
output_y = np.array([-3,-1,1,3,5,7],dtype=float)
```

```python
for i,c in enumerate(input_x):
    print(c,output_y[i])
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1,input_shape=[1])
])
```

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

```python
model_data = model.fit(input_x,output_y,epochs=500)
```

```python
model_data.history
```

```python
import matplotlib.pyplot as plt

plt.plot(model_data.history['loss'])
plt.show()
```

```python
model.predict([-1])
```

```python
#Valores conocidos

print('x:-1 ------',normalRule(-1),model.predict([-1]))
print('x: 0 ------',normalRule(0),model.predict([0]))
print('x: 1 ------',normalRule(1),model.predict([1]))
print('x: 2 ------',normalRule(2),model.predict([2]))
```

```python
#Valores desconocidos

print('x:10 ------',normalRule(10),model.predict([10]))
print('x:45 ------',normalRule(45),model.predict([45]))
```
